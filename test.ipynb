```python
import torch
import numpy as np
from scipy.linalg import hadamard 
dd = 12
DD = 780
LL = 1024
padding_size = LL-DD
a = np.random.randn(DD)
b1 = np.random.randn(512)
b1[b1>0]=1
b1[b1<0]=-1
b2 = np.random.randn(512)
b2[b2>0]=1
b2[b2<0]=-1

h = hadamard(LL)/np.sqrt(LL)
a_pad = np.append(a,np.zeros(padding_size))
subspace = a_pad@h@h*np.sqrt(DD/dd)
subspace = subspace[:dd]
print(np.linalg.norm(a),np.linalg.norm(subspace))
recover = np.append(subspace,torch.zeros(LL-dd))
DD_recover = recover@h@h
DD_recover = DD_recover[:DD]
print(np.linalg.norm(a),np.linalg.norm(DD_recover))
# print(a-DD_recover)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from scipy.linalg import hadamard
import numpy as np
d = 30
D = 1024
a = torch.zeros(dd)
h = hadamard(D)
h = torch.from_numpy(h)
h = h.type(torch.float)
padding_size = D-d

for i in range(10):
  a += torch.randn(d)
  a_pad = F.pad(a,pad = (0,padding_size),value=0.0,mode='constant')
  print(torch.count_nonzero(torch.matmul(a_pad,h)))
```
